# time and datetime 

Ժամանակի և տարեթվերի հետ աշխատելիս հաճախ կհանդիպենք epoch բառը։ Epoch-ը այն ժամանակի սկիզբն է, որից սկսված համակարգչային օպերացիոն համակարգերը հաշվում են ժամանակը։ Epoch-ը սկսվում է 1970 թվականի հունվարի 1-ից։

````python
import time
````

````python
print(time.time())
````

Ավելի ընթեռնելի դարձնելու համար կարող ենք օգտագործել **strftime()** (string from time) մեթոդը։ Առաջին արգումենտով ֆորմատավորում ենք ժամանակը։

Տվյալ մեթոդի և **time()** մոդուլի մասին ավելի մանրամասն կարող եք կարդալ այստեղ՝
https://docs.python.org/3/library/time.html

````python
print(time.gmtime()[0])
print(time.gmtime().tm_year)

print(time.localtime())

print(time.ctime())
````

````python
time.strftime('%B, %x, %X', time.gmtime())
````

time-ը կարելի է օգտագործել որպես վայրկյանաչափ։ Եկեք գրենք մի փոքրիկ խաղ, որը ստուգում է մեր ռեակցիայի ժամանակը։ 

````python
import random

input('Press enter to start...')

random_num = random.randint(1, 6)
time.sleep(random_num)

start_time = time.time()

input('Press Enter, quickly!')

end_time = time.time()

print('Your reaction time was {}'.format(end_time - start_time))
````

````python
import random

input('Press enter to start...')

random_num = random.randint(1, 6)
time.sleep(random_num)

start_time = time.perf_counter()

input('Press Enter, quickly!')

end_time = time.perf_counter()

print('Your reaction time was {}'.format(end_time - start_time))
````

````python
import random

input('Press enter to start...')

random_num = random.randint(1, 6)
time.sleep(random_num)

start_time = time.monotonic()

input('Press Enter, quickly!')

end_time = time.monotonic()

print('Your reaction time was {}'.format(end_time - start_time))
````

time.time()-ը որպես վայրկյանաչափ օգտագործելու խնդիրն այն է, որ եթե տվյալ երկրում ժամերը առաջ կամ հետ տանեն մեր խաղալու ընթացքում, ապա ռեակցիայի ժամանակը կլինի մեկ ժամից ավել։ Դրա համար կարող ենք օգտագործել **perf_counter()** մեթոդը, որն իրական ժամանակից անկախ ճշգրիտ վայրկյանաչափ է։

**timedelta**-ն թույլ է տալիս համեմատել տարբեր ամսաթվեր։

````python
queen_elizabeth_birthday = datetime(1926, 4, 21, 17, 55)
putin_birthday = datetime(1952, 10, 7, 18, 30)
````

````python
tz_london = pytz.timezone('Europe/London')
tz_moscow = pytz.timezone('Europe/Moscow')

elizabeth = tz_london.localize(queen_elizabeth_birthday)
print(elizabeth)
putin = tz_moscow.localize(putin_birthday)
````

````python
print(type(elizabeth))
````

````python
delta = elizabeth - putin
print(delta)
print(math.fabs(delta.days))
````

# random

Համակարգչի մեջ թվերը կատարելապես պատահական չեն։ Պատահական թվերի գեներատորը օգտագործում է այսպես կոչված seed, որի հիման վրա վերադարձնում է պատահական թիվը։ seed-ը թիվ է։ Եթե մենք բացահայտ կերպով այն չենք հայտարարում, ապա գեներատորը որպես seed օգտագործում է համակարգչի ժամացույցը։

**random()** մեթոդը վերադարձնում է պատահական թիվ [0, 1) միջակայքում։

````python
random_number = random.random()
random_number
````

**randint(a, b)** մեթոդը վերադարձնում է ամբողջ թիվ [a, b] միջակայքում:

````python
random_int = random.randint(1, 100)
random_int
````

Եթե **seed**-ը բացահայտ կերպով նշանակենք, ապա միշտ կստանանք նույն պատահական թվերի հաջորդականությունը։

````python
random.seed(12837192)

print(random.randint(1, 100))
````

**randrange(a, b, step)**-ն վերադարձնում է թիվ [a, b) միջակայքում։ Այն նաև կարող է վերցնել երրորդ արգումենտ, որը սահմանում է տիրույթի քայլը։

````python
random_int_2 = random.randrange(0, 100, 3)
random_int_2
````

**shuffle()** մեթոդը պատահականորեն խառնում է փոխանցված հաջորդականությունը

````python
lst = [i for i in range(0, 20)]
print(lst)
random.shuffle(lst)
print(lst)
````

random մոդուլի հետ ավելի մանրամասն կարող եք ծանոթանալ այստեղ․

```

r - read
```
