### Copying a list

Լիստը copy անելու համար կարող է տրամաբանական թվալ հետևյալ օրինակը։

```python
list_a = [1, 2, 3]
```

```python
list_b = list_a
```

Մենք հայտարարեցինք **list_a** և **list_b**, և երկրորդ լիստը նշանակեցինք առաջին լիստով։ Հիմա որոշակի գործողություններ կատարենք առաջին լիստի հետ (օրինակ **append**, **pop**):

```python
list_a
```

```python
list_b
```

```python
list_b.append('K')
```

```python
print('list_b', list_b)
print('list_a', list_a)
```

```python
print(hex(id(list_a)))
print(hex(id(list_b)))
```

ինչպես տեսնում ենք, երկրորդ լիստը նույնպես փոխվեց։ Երբ մենք հայտարարեցինք **list_b = list_a**, մենք հիշողության նույն հասցեն վերագրեցինք նաև list_b-ին։ Այժմ մենք ունենք երկու փոփոխական, որոնք հղում են կատարում հիշողության նույն հատվածին, հետևաբար փոխելով մեկը՝ փոխում ենք մյուսը։

Դրանից խուսափելու համար, կարող ենք օգտագործել **list()** ֆունկցիան։ Դրանով մենք ստեղծում ենք նոր փոփոխական, որը ստանում է այն արժեքները, ինչ-որ կար առաջին լիստ-ում

```python
list_c = list(list_a)
```

```python
list_c
```

```python
print(hex(id(list_a)))
print(hex(id(list_c)))
```

```python
list_c.pop(-1)
list_c
```

```python
list_a
```
